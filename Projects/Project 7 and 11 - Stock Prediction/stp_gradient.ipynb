```python
dataset = dataset.dropna(how = "any" )
```

```python
dataset = dataset.drop(columns="Volume")
```

```python
print(dataset.head())
def PlotData(features):
    plt.figure(figsize= (20, 5))    
    pltNum = 1
    for mem in features:
        plt.subplot(1, 4, pltNum)
        plt.style.use('seaborn-whitegrid')
        plt.grid(True)
        plt.title('Regplot Plot for '+ str(mem))
        sns.regplot(data = dataset, x = mem, y = 'Open' , color = 'green')
        pltNum += 1
    plt.show()
```

```python
PlotData(['High', 'Low', 'Close', 'Adj Close'])
```
